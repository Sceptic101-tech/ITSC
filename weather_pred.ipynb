```python
df = pd.read_csv('datasets/train_df.csv')
coordinate_df = pd.DataFrame(coordinates)
```

```python
df.head()
```

```python
#Соединяем два датафрейма
df = pd.merge(df, coordinate_df, how='outer', on='Location')
```

```python
#df['RainTomorrow'].plot(kind='hist')
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
category_list = ['Location', 'RainToday', 'RainTomorrow', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
```

```python
#Отмечаем, было ли значние NaN
for column in df.columns:
    if df[column].isna().sum() > 0:
        df['was_NA_' + column] = 0
        df.loc[df[column].isna() == True, 'was_NA_' + column] = 1
```

# Заполнение пропусков простой вставкой занчения

```python
comment = '''
#Категориальные заполняем Unknown, а числовые - медианой
df = df.apply(lambda x: x.fillna('Unknown') if x.name in category_list else x.fillna(x.median()))
'''
```

# Encoding

```python
encoder = OrdinalEncoder(dtype=np.float64)
df[category_list] = encoder.fit_transform(df[category_list])

df = df.rename(columns={'Location' : 'loc_id'})

#Так получается, что столбцу RainToday OrdinalEncoder присваивает в случае No = 0, Yes = 2, Unknown = 1. Меняем значение Unknown на -1, а Yes на 1
df.loc[df.RainToday == 1, 'RainToday'] = -1
df.loc[df.RainToday == 2, 'RainToday'] = 1
```

```python
df['year'] = df['Date'].apply(lambda x: x.year)
df['month'] = df['Date'].apply(lambda x: x.month)
df['day_of_week'] = df['Date'].apply(lambda x: x.dayofweek)
df['week_of_year'] = df['Date'].apply(lambda x: x.weekofyear)
df = df.sort_values(['loc_id', 'Date']).reset_index().drop('index', axis=1)
```

# Заполнение пропусков методом KNN

```python
without_date = df.columns.values[1:] #Необходимо для работы скейлера и KNN
num_features = df.select_dtypes(include=['int64', 'float64']).columns.values

scaler = MinMaxScaler(feature_range=(0, 1))
df[num_features] = scaler.fit_transform(df[num_features])
```

```python
imputer=KNNImputer(n_neighbors=2, weights='uniform')
imputer.fit(df[without_date])

df[without_date] = pd.DataFrame(imputer.transform(df[without_date]), index=df[without_date].index, columns = df[without_date].columns)
```

# Поиск разницы между двумя ближайшими записями

```python
#Разница между двумя ближайшими записями в датасете для каждого города

#Создаем список с датами для каждого loc_id
gap_days = df.groupby('loc_id').Date.apply(list)

#Дублируем для каждого loc_id дату начала наблюдений, чтобы при вычислении разности между днями массив остался той же длины
gap_days = gap_days.map(lambda x: [x[0]] + x).apply(np.diff).map(lambda x: [xi.days for xi in x])

#Превращаем все в серию, отсортированную по loc_id
gap_days = pd.Series(np.concatenate(gap_days.values))
```

```python
df['delta_days_of_records'] = gap_days
df['days_without_rain'] = 0
```

```python
for i in range(1,df.shape[0]):
    if (df.loc[i,'delta_days_of_records'] > 2) or (df.loc[i, 'was_NA_RainToday'] == 1):#((df.loc[i,'RainToday'] != 0) and (df.loc[i,'RainToday'] != 2)):
        df.loc[i, 'days_without_rain'] = -10;
    else:
        if df.loc[i,'RainToday'] == 0: #Если дождь
            df.loc[i, 'days_without_rain'] = 0
        else:
            if df.loc[i, 'loc_id'] == df.loc[i-1, 'loc_id']:
                df.loc[i, 'days_without_rain'] = df.loc[i-1, 'days_without_rain'] + 1
```

```python
df.head(10)
```

# FIT -> PREDICT

```python
X = df.drop(['RainTomorrow', 'Date', 'delta_days_of_records'], axis=1)
y = df['RainTomorrow']

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
print(f"Результаты: {scores.mean():.4f} ± {scores.std():.4f}")
```
